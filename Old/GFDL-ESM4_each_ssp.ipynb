```python
# Define Earth's radius in meters
R = 6378137

# Latitude and longitude arrays (in degrees)
lon_tas = datasets_tas_historical.coords['lon']
lat_tas = datasets_tas_historical.coords['lat']

# Convert to radians
lon_radians_tas = np.radians(lon_tas)
lat_radians_tas = np.radians(lat_tas)

# Calculate the differences in longitude and latitude
delta_lon_tas = np.diff(lon_radians_tas)[0]  # Assuming regular grid
delta_lat_tas = np.diff(lat_radians_tas).mean()  # Take the mean for approximate spacing

# Calculate area for each grid cell
area_tas = np.zeros((len(lat_tas), len(lon_tas)))  # Initialize area array to match full grid

for i in range(len(lat_tas)):
    # Latitude bounds (extend to midpoints for full grid)
    if i == 0:  # Southernmost edge
        phi1_tas = lat_radians_tas[i] - delta_lat_tas / 2
        phi2_tas = lat_radians_tas[i] + delta_lat_tas / 2
    elif i == len(lat_tas) - 1:  # Northernmost edge
        phi1_tas = lat_radians_tas[i] - delta_lat_tas / 2
        phi2_tas = lat_radians_tas[i] + delta_lat_tas / 2
    else:  # Interior latitudes
        phi1_tas = lat_radians_tas[i] - delta_lat_tas / 2
        phi2_tas = lat_radians_tas[i] + delta_lat_tas / 2

    # Calculate area for this latitude band
    area_tas[i, :] = (R**2 * delta_lon_tas * (np.sin(phi2_tas) - np.sin(phi1_tas)))

# Convert to square kilometers
area_km2_tas = area_tas / 1e6

# Print or save the area array
print("Area array (in km^2):", area_km2_tas)
print("Shape of area array:", area_km2_tas.shape)
```

```python
def prepare_panel_th_for_plots_and_mlr(
    datasets_fwi, datasets_tasmax, datasets_pr, datasets_hurs, datasets_sfcWind,
    labels, areacella_fwi, areacella_tas,
    lat_bounds=(5.5, 20.5), lon_bounds=(97.5, 105.5),
    fwi_is_yearly=True
):
    """
    เตรียม DataFrame รายปีของไทย สำหรับใช้วิเคราะห์/ทำกราฟ/ทำ MLR:
      - Slice พื้นที่ไทย
      - FWI: ใช้รายปี (ไม่ resample ถ้า fwi_is_yearly=True)
      - tasmax/pr/hurs/sfcWind: resample รายปีด้วย mean
      - น้ำหนักพื้นที่: FWI ใช้ areacella_fwi['area'], ตัวอื่นใช้ areacella_tas['area']
      - จัดเวลาให้ตรงกันกับ FWI ในแต่ละ scenario
      - แปลงหน่วย: tasmax K→°C, pr kg m^-2 s^-1 → mm/day (คูณ 86400 หลังเฉลี่ยรายปี)
    คืนค่า: panel_df พร้อมคอลัมน์ ['time','scenario','FWI','tasmax','pr','hurs','sfcWind']
    """

    lat_slice = slice(*lat_bounds)
    lon_slice = slice(*lon_bounds)

    # จัดการ areacella: ตัดมิติเวลาออกหากมี และตัดเฉพาะไทย
    area_fwi = areacella_fwi
    #Right here it will pass if since there is no time dimension in areacella
    if 'time' in area_fwi.dims:
        area_fwi = area_fwi.isel(time=0)
    area_fwi = area_fwi['area'].sel(lat=lat_slice, lon=lon_slice)

    area_tas = areacella_tas
    if 'time' in area_tas.dims:
        area_tas = area_tas.isel(time=0)
    area_tas = area_tas['area'].sel(lat=lat_slice, lon=lon_slice)

    rows = []

    for ds_fwi, ds_tas, ds_pr, ds_hurs, ds_wind, label in zip(
        datasets_fwi, datasets_tasmax, datasets_pr, datasets_hurs, datasets_sfcWind, labels
    ):
        # 1) ตัดพื้นที่ไทย
        fwi_th   = ds_fwi .sel(lat=lat_slice, lon=lon_slice)
        tas_th   = ds_tas .sel(lat=lat_slice, lon=lon_slice)
        pr_th    = ds_pr  .sel(lat=lat_slice, lon=lon_slice)
        hurs_th  = ds_hurs.sel(lat=lat_slice, lon=lon_slice)
        wind_th  = ds_wind.sel(lat=lat_slice, lon=lon_slice)

        # 2) ตัวแปร
        fwi_da   = fwi_th['fwisa']                        # FWI (รายปีอยู่แล้วในหลายชุด)
        tas_da   = tas_th['tasmax']                       # Kelvin
        pr_da    = pr_th['pr']                            # kg m^-2 s^-1 (≈ mm/s)
        hurs_da  = hurs_th['hurs']                        # %
        wind_da  = wind_th['sfcWind']                     # m/s

        # 3) ทำรายปี
        fwi_y    = fwi_da if fwi_is_yearly else fwi_da.resample(time='YE').mean()
        tas_y_K  = tas_da.resample(time='YE').mean()
        pr_y     = pr_da .resample(time='YE').mean()      # จะคูณ 86400 เป็น mm/day
        hurs_y   = hurs_da.resample(time='YE').mean()
        wind_y   = wind_da.resample(time='YE').mean()

        # 4) จัดเวลาให้ตรงกับ FWI ของ scenario เดียวกัน (ใกล้สุด)
        tas_y_K  = tas_y_K .sel(time=fwi_y['time'], method='nearest')
        pr_y     = pr_y    .sel(time=fwi_y['time'], method='nearest')
        hurs_y   = hurs_y  .sel(time=fwi_y['time'], method='nearest')
        wind_y   = wind_y  .sel(time=fwi_y['time'], method='nearest')

        # 5) เฉลี่ยถ่วงน้ำหนักพื้นที่ (FWI ใช้ area_fwi, ตัวอื่นใช้ area_tas)
        fwi_mean    = fwi_y  .weighted(area_fwi).mean(dim=['lat','lon'])
        tas_yK_mean = tas_y_K.weighted(area_tas).mean(dim=['lat','lon'])
        pr_mean     = pr_y   .weighted(area_tas).mean(dim=['lat','lon'])
        hurs_mean   = hurs_y .weighted(area_tas).mean(dim=['lat','lon'])
        wind_mean   = wind_y .weighted(area_tas).mean(dim=['lat','lon'])

        # 6) แปลงหน่วย
        tas_mean_C = tas_yK_mean - 273.15          # °C
        pr_mean_md = pr_mean * 86400.0             # mm/day (mean over the year)

        # 7) รวบรวมเป็น DataFrame ต่อ scenario
        years = pd.Index(fwi_mean.indexes['time'].year, name='time')
        df = pd.DataFrame({
            'time'   : years.values.astype(int),
            'FWI'    : fwi_mean.values,
            'tasmax' : tas_mean_C.values,
            'pr'     : pr_mean_md.values,
            'hurs'   : hurs_mean.values,
            'sfcWind': wind_mean.values,
            'scenario': label
        })
        rows.append(df)

    panel_df = pd.concat(rows, ignore_index=True)

    # 8) ให้แน่ใจว่าเป็นตัวเลข และเอา inf/NaN ออก (เตรียมพร้อมสำหรับ plot/MLR)
    for c in ['FWI','tasmax','pr','hurs','sfcWind']:
        panel_df[c] = pd.to_numeric(panel_df[c], errors='coerce')
    panel_df.replace([np.inf, -np.inf], np.nan, inplace=True)
    panel_df.dropna(subset=['FWI','tasmax','pr','hurs','sfcWind'], inplace=True)
    panel_df['time'] = panel_df['time'].astype(int)

    return panel_df

panel_df_th = prepare_panel_th_for_plots_and_mlr(
    datasets_fwi, datasets_temp, datasets_pr, datasets_hurs, datasets_sfcWind,
    labels_all_scenarios, areacella, areacella_tas,  # FWI weights, TAS weights
    fwi_is_yearly=True  # หาก FWI ของคุณเป็นรายเดือน ให้ตั้ง False
)

# ตอนนี้ panel_df_th พร้อมแล้วสำหรับ:
# - ทำกราฟ FWI vs tasmax/pr/hurs/sfcWind ตาม scenario
# - หรือรัน MLR ด้วย statsmodels ภายหลัง
print(panel_df_th.head())
print(panel_df_th.groupby('scenario').size())
```

```python
def apply_ols_prediction(df, intercept, beta_tasmax, beta_pr, beta_hurs, beta_sfcWind):
    """
    Apply OLS regression coefficients to predict FWI.

    Parameters
    ----------
    df : pandas.DataFrame
        Input DataFrame containing columns: 'tasmax', 'pr', 'hurs', 'sfcWind'
    intercept : float
        Regression intercept
    beta_tasmax, beta_pr, beta_hurs, beta_sfcWind : float
        Beta coefficients for respective variables

    Returns
    -------
    df : pandas.DataFrame
        Copy of input DataFrame with a new column 'FWI_pred'
    """
    df = df.copy()  # keep original intact
    df["FWI_pred"] = (
        intercept
        + beta_tasmax * df["tasmax"]
        + beta_pr * df["pr"]
        + beta_hurs * df["hurs"]
        + beta_sfcWind * df["sfcWind"]
    )
    return df

# Historical
hist_df = apply_ols_prediction(
    df=panel_df_th[panel_df_th["scenario_short"] == "Historical"],
    intercept=-45.7475,
    beta_tasmax=0.2685,
    beta_pr=-2.1548,
    beta_hurs=0.3562,
    beta_sfcWind=14.6480
)

# SSP1-1.9
ssp119_df = apply_ols_prediction(
    df=panel_df_th[panel_df_th["scenario_short"] == "SSP1-1.9"],
    intercept=-71.1223,
    beta_tasmax=2.2981,
    beta_pr=-1.3768,
    beta_hurs=0.3932,
    beta_sfcWind=1.3501
)

# SSP1-2.6
ssp126_df = apply_ols_prediction(
    df=panel_df_th[panel_df_th["scenario_short"] == "SSP1-2.6"],
    intercept=-46.6780,
    beta_tasmax=1.1322,
    beta_pr=-0.6685,
    beta_hurs=0.2101,
    beta_sfcWind=7.9084
)

# SSP2-4.5
ssp245_df = apply_ols_prediction(
    df=panel_df_th[panel_df_th["scenario_short"] == "SSP2-4.5"],
    intercept=-29.0741,
    beta_tasmax=0.3363,
    beta_pr=-4.3598,
    beta_hurs=0.3445,
    beta_sfcWind=12.4733
)

# SSP3-7.0
ssp370_df = apply_ols_prediction(
    df=panel_df_th[panel_df_th["scenario_short"] == "SSP3-7.0"],
    intercept=-76.8222,
    beta_tasmax=1.4443,
    beta_pr=-4.4467,
    beta_hurs=0.6069,
    beta_sfcWind=10.8384
)

# SSP5-8.5
ssp585_df = apply_ols_prediction(
    df=panel_df_th[panel_df_th["scenario_short"] == "SSP5-8.5"],
    intercept=-94.3122,
    beta_tasmax=2.0384,
    beta_pr=-3.0760,
    beta_hurs=0.6232,
    beta_sfcWind=8.1957
)
```

```python
# =========================
# OLS per-scenario (HC1) + standardized betas + tidy tables
# =========================
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
from scipy.stats import zscore

# ---- 0) Settings ----
RESPONSE   = "FWI"
PREDICTORS = ["tasmax", "pr", "hurs", "sfcWind"]   # swap "hurs" -> "h2o" if you prefer

# If you want shorter scenario names, you can map them first (optional):
scenario_map = {
    "Historical to Present (1850-2015)": "Historical",
    'SSP1-1.9 ("Sustainability - Very Low Emissions")': "SSP119",
    'SSP1-2.6 ("Sustainability - Low Emissions with Strong Mitigation")': "SSP126",
    'SSP2-4.5 ("Middle of the Road - Moderate Emissions")': "SSP245",
    'SSP3-7.0 ("Regional Rivalry - High Emissions")': "SSP370",
    'SSP4-3.4 ("Inequality - Intermediate Emissions")': "SSP434",
    'SSP4-6.0 ("Inequality - High Emissions")': "SSP460",
    'SSP5-3.4 ("Fossil-Fueled Development - Moderate Emissions")': "SSP534",
    'SSP5-8.5 ("Fossil-Fueled Development - High Emissions, No Climate Policy")': "SSP585",
}
panel_df_th["scenario_short"] = panel_df_th["scenario"].map(scenario_map).fillna(panel_df_th["scenario"])

scenarios = list(panel_df_th["scenario_short"].unique())

# Containers for combined outputs
per_sc_models       = {}         # scenario -> fitted model (unstandardized)
per_sc_models_beta  = {}         # scenario -> fitted model on z-scored data
coef_tables         = []         # long-form rows (per scenario)
beta_tables         = []         # standardized betas (per scenario)
metrics_rows        = []         # R2 etc.

for sc in scenarios:
    df_sc = panel_df_th.loc[panel_df_th["scenario_short"] == sc, [RESPONSE] + PREDICTORS].dropna()
    if df_sc.shape[0] < len(PREDICTORS) + 2:
        # too few rows to reliably fit; skip
        print(f"Skip {sc}: not enough rows ({df_sc.shape[0]})")
        continue

    # --- 1) Unstandardized OLS with HC1 ---
    formula_sc = f"{RESPONSE} ~ " + " + ".join(PREDICTORS)  # no FE term; this is per-scenario
    model_sc = smf.ols(formula_sc, data=df_sc).fit(cov_type="HC1")
    per_sc_models[sc] = model_sc

    # --- 2) Standardized betas (z-score within scenario) ---
    df_sc_std = df_sc.copy()
    cols_to_z = [RESPONSE] + PREDICTORS
    df_sc_std[cols_to_z] = df_sc_std[cols_to_z].apply(zscore)
    model_sc_beta = smf.ols(formula_sc, data=df_sc_std).fit(cov_type="HC1")
    per_sc_models_beta[sc] = model_sc_beta

    # --- 3) Collect tidy coef table (unstandardized) ---
    ct = pd.DataFrame({
        "scenario": sc,
        "term": model_sc.params.index,
        "coef": model_sc.params.values,
        "HC1_se": model_sc.bse.values,
        "t": model_sc.tvalues.values,
        "pvalue": model_sc.pvalues.values,
    })
    ct["CI_low"]  = ct["coef"] - 1.96 * ct["HC1_se"]
    ct["CI_high"] = ct["coef"] + 1.96 * ct["HC1_se"]
    coef_tables.append(ct)

    # --- 4) Collect standardized betas (only predictors) ---
    bt = pd.Series(
        {k: v for k, v in model_sc_beta.params.items() if k in PREDICTORS},
        name=sc
    )
    beta_tables.append(bt)

    # --- 5) Metrics per scenario ---
    metrics_rows.append(pd.Series({
        "scenario": sc,
        "n": df_sc.shape[0],
        "k": len(PREDICTORS),
        "R2": model_sc.rsquared,
        "Adj_R2": model_sc.rsquared_adj,
        "F": model_sc.fvalue if np.isfinite(model_sc.fvalue) else np.nan
    }))

# ---- Combine outputs ----
coef_table_all = pd.concat(coef_tables, ignore_index=True) if coef_tables else pd.DataFrame()
beta_table     = pd.DataFrame(beta_tables) if beta_tables else pd.DataFrame()
metrics_table  = pd.DataFrame(metrics_rows).set_index("scenario").sort_index() if metrics_rows else pd.DataFrame()

# ---- Pretty prints ----
print("\n=== OLS per-scenario: coefficients (HC1) ===")
if not coef_table_all.empty:
    print(coef_table_all.round(4))
else:
    print("No results (check data filtering).")

print("\n=== Standardized (beta) coefficients by scenario ===")
if not beta_table.empty:
    print(beta_table.round(3))
else:
    print("No beta results (check data filtering).")

print("\n=== Model metrics per scenario ===")
if not metrics_table.empty:
    print(metrics_table.round(3))
else:
    print("No metrics results (check data filtering).")
```

```python
# ========= OVERLAID, NORMALIZED HISTOGRAMS with SCALING =========
# Compare distributions fairly across scenarios (and even across variables)
# Author: you + ChatGPT :)

import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from pathlib import Path

# ===================== CONFIG =====================
df = panel_df_th.copy()   # your DataFrame

# Precipitation input unit: "auto" | "kg m-2 s-1" | "mm/day" | "mm/yr"
PR_UNIT_IN = "mm/yr"

DAYS_PER_YEAR = 365.2422
SECONDS_PER_DAY = 86400.0
SECONDS_PER_YEAR = SECONDS_PER_DAY * DAYS_PER_YEAR

# Variables to include (we'll create pr_mm_yr)
vars_to_plot = ["FWI", "tasmax", "pr_mm_yr", "hurs", "sfcWind"]

palette_all = {
    "Historical": "#000000",
    "SSP1-1.9":   "#E69F00",
    "SSP1-2.6":   "#56B4E9",
    "SSP2-4.5":   "#009E73",
    "SSP3-7.0":   "#882255",
    "SSP4-3.4":   "#E6D600",
    "SSP4-6.0":   "#D55E00",
    "SSP5-3.4":   "#0072B2",
    "SSP5-8.5":   "#CC79A7",
}

labels_map = {
    "FWI": "FWI (dimensionless)",
    "tasmax": "tasmax (°C)",
    "pr_mm_yr": "Precipitation (mm/year)",
    "hurs": "Relative humidity (%)",
    "sfcWind": "Surface wind (m/s)",
}

# Trim extreme upper tail uniformly across scenarios
upper_pct = 99.5  # set None to disable trimming

# If you want to pre-transform some skewed variables before scaling (e.g., log10)
# (applied to positive values only, AFTER trimming)
pretransform_log10_vars = set()  # e.g., {"pr_mm_yr"}

# ---- How to compare values across variables ----
# "none"     : raw values (best when comparing scenarios of the SAME variable)
# "z"        : z-score (mean/std)
# "robust_z" : robust z ((x-median) / (IQR/1.349)) - good if outliers/skew
# "minmax"   : scale to [0, 1]
scale_mode = "robust_z"

# ---- How to normalize the histograms ----
# "pdf"      : area under histogram = 1  (y = probability density; can exceed 1)
# "fraction" : sum of bar heights = 1    (y = fraction of samples; bounded <= 1)
normalize_mode = "pdf"

# ---- Saving options ----
OUTPUT_DIR = Path("fig_distros_GFDL_OVERLAID_normalized")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
SAVE_FORMATS = ["png", "pdf"]
DPI = 200
TRANSPARENT = False

# ===================== STYLE MAPS =====================
hatch_map = {
    "Historical": "",
    "SSP1-1.9":  "//",
    "SSP1-2.6":  "\\\\",
    "SSP2-4.5":  "xx",
    "SSP3-7.0":  "..",
    "SSP4-3.4":  "++",
    "SSP4-6.0":  "**",
    "SSP5-3.4":  "oo",
    "SSP5-8.5":  "--",
}

line_map_str = {
    "Historical": "solid",
    "SSP1-1.9":  "dashed",
    "SSP1-2.6":  "dotted",
    "SSP2-4.5":  "dashdot",
    "SSP3-7.0":  "dashed",
    "SSP4-3.4":  "dotted",
    "SSP4-6.0":  "dashdot",
    "SSP5-3.4":  "dashed",
    "SSP5-8.5":  "dotted",
}

def _safe_linestyle(scen):
    return line_map_str.get(scen, "solid")

# ===================== HELPERS =====================
def canonical_scenario(s):
    if pd.isna(s): return s
    s = str(s)
    if "Historical" in s or "1850-2015" in s:
        return "Historical"
    m = re.search(r"SSP[1-5]-\d(?:\.\d)?", s)
    return m.group(0) if m else s.strip()

def pr_to_mm_per_year(series, assume="auto"):
    x = pd.to_numeric(series, errors="coerce").astype(float)
    if assume == "kg m-2 s-1":
        return x * SECONDS_PER_YEAR
    elif assume == "mm/day":
        return x * DAYS_PER_YEAR
    elif assume == "mm/yr":
        return x
    else:
        q95 = np.nanpercentile(x, 95)
        return x * SECONDS_PER_YEAR if (np.isfinite(q95) and q95 < 1.0) else x * DAYS_PER_YEAR

def common_bins(values, max_bins=120):
    edges = np.histogram_bin_edges(values, bins="fd")
    if len(edges) > max_bins:
        edges = np.linspace(edges.min(), edges.max(), max_bins)
    return edges

def fit_scaler(values: np.ndarray, mode: str):
    """Fit a scaler on pooled (trimmed/transformed) values and return a transform(x) fn."""
    v = values[np.isfinite(values)]
    if mode == "none":
        return lambda x: x, {"label_suffix": ""}
    if mode == "z":
        mu = np.nanmean(v)
        sd = np.nanstd(v, ddof=0)
        sd = sd if sd > 0 else np.nan
        return (lambda x: (x - mu) / sd if np.isfinite(sd) else x*0), {"label_suffix": " [z-score]"}
    if mode == "robust_z":
        med = np.nanmedian(v)
        q1, q3 = np.nanpercentile(v, [25, 75])
        iqr = q3 - q1
        sd_r = (iqr / 1.349) if iqr > 0 else np.nan
        return (lambda x: (x - med) / sd_r if np.isfinite(sd_r) else x*0), {"label_suffix": " [robust z]"}
    if mode == "minmax":
        mn, mx = np.nanmin(v), np.nanmax(v)
        rng = mx - mn
        rng = rng if rng > 0 else np.nan
        return (lambda x: (x - mn) / rng if np.isfinite(rng) else x*0), {"label_suffix": " [min–max]"}
    raise ValueError("Unknown scale_mode")

def pretransform(values: np.ndarray, varname: str):
    """Optional log10 transform BEFORE scaling (for skewed positives)."""
    if varname in pretransform_log10_vars:
        vals = values.copy()
        vals = vals[~np.isnan(vals)]
        # Keep only positive for log
        return np.log10(values, where=(values > 0), out=np.full_like(values, np.nan))
    return values

# ===================== PREP =====================
scenario_col = "scenario_short" if "scenario_short" in df.columns else "scenario"
df["scenario"] = df[scenario_col].apply(canonical_scenario)

canonical_order = ["Historical","SSP1-1.9","SSP1-2.6","SSP2-4.5","SSP3-7.0","SSP4-3.4","SSP4-6.0","SSP5-3.4","SSP5-8.5"]
present = [s for s in canonical_order if s in set(df["scenario"].dropna().unique())]
df = df[df["scenario"].isin(present)].copy()
df["scenario"] = pd.Categorical(df["scenario"], categories=present, ordered=True)

# tasmax K -> °C if needed
if pd.to_numeric(df["tasmax"], errors="coerce").median() > 100:
    df["tasmax"] = pd.to_numeric(df["tasmax"], errors="coerce") - 273.15

# pr -> mm/yr (clip small negatives)
df["pr_mm_yr"] = pr_to_mm_per_year(df["pr"], assume=PR_UNIT_IN).clip(lower=0)

# Ensure numeric + drop NaNs on used vars
for col in vars_to_plot:
    df[col] = pd.to_numeric(df[col], errors="coerce")
df = df.dropna(subset=vars_to_plot + ["scenario"])

# Color map limited to present scenarios
palette = {k: palette_all[k] for k in present if k in palette_all}

# ===================== PLOT & SAVE =====================
for var in vars_to_plot:
    # --- pooled raw values for trimming threshold ---
    pooled_raw = df[var].to_numpy()
    pooled_raw = pooled_raw[np.isfinite(pooled_raw)]
    if pooled_raw.size == 0:
        continue

    if upper_pct is not None:
        cap = np.nanpercentile(pooled_raw, upper_pct)
    else:
        cap = np.inf

    # --- apply uniform trimming + optional pre-transform BEFORE fitting scaler ---
    pooled_trim = pooled_raw[pooled_raw <= cap]
    pooled_trim = pretransform(pooled_trim, var)

    # Fit scaler on pooled, trimmed, pre-transformed
    transform, meta = fit_scaler(pooled_trim, scale_mode)

    # Build pooled SCALED values (used for bins)
    pooled_scaled_list = []
    for scen in present:
        vals = df.loc[df["scenario"] == scen, var].to_numpy()
        vals = vals[np.isfinite(vals)]
        vals = vals[vals <= cap]
        vals = pretransform(vals, var)
        vals = transform(vals)
        pooled_scaled_list.append(vals)
    pooled_scaled = np.concatenate([v for v in pooled_scaled_list if v.size > 0], axis=0)
    if pooled_scaled.size == 0:
        continue

    bins = common_bins(pooled_scaled)

    fig, ax = plt.subplots(figsize=(9, 6))

    # -------- Filled bars with hatches (normalized as requested) --------
    for scen, vals_scaled in zip(present, pooled_scaled_list):
        vals_scaled = vals_scaled[np.isfinite(vals_scaled)]
        if vals_scaled.size == 0:
            continue

        if normalize_mode == "pdf":
            # area-normalized (probability density)
            _, _, patches = ax.hist(
                vals_scaled, bins=bins, density=True, alpha=0.18,
                label=None, color=palette.get(scen), edgecolor="none", zorder=1
            )
        elif normalize_mode == "fraction":
            # heights sum to 1 (fraction of samples)
            weights = np.ones_like(vals_scaled) / vals_scaled.size
            _, _, patches = ax.hist(
                vals_scaled, bins=bins, weights=weights, alpha=0.18,
                label=None, color=palette.get(scen), edgecolor="none", zorder=1
            )
        else:
            raise ValueError("normalize_mode must be 'pdf' or 'fraction'")

        hatch = hatch_map.get(scen, "")
        if hatch:
            for p in patches:
                p.set_hatch(hatch)

    # -------- Step outlines on top --------
    for scen, vals_scaled in zip(present, pooled_scaled_list):
        vals_scaled = vals_scaled[np.isfinite(vals_scaled)]
        if vals_scaled.size == 0:
            continue

        if normalize_mode == "pdf":
            ax.hist(
                vals_scaled, bins=bins, density=True, histtype="step",
                linewidth=2.0, color=palette.get(scen),
                linestyle=_safe_linestyle(scen), zorder=2
            )
        else:  # "fraction"
            weights = np.ones_like(vals_scaled) / vals_scaled.size
            ax.hist(
                vals_scaled, bins=bins, weights=weights, histtype="step",
                linewidth=2.0, color=palette.get(scen),
                linestyle=_safe_linestyle(scen), zorder=2
            )

    # -------- Axes labels/titles --------
    ylab = "Probability density" if normalize_mode == "pdf" else "Fraction of samples"
    xlab_suffix = meta.get("label_suffix", "")
    # Clarify if we applied log10 pretransform
    prelog_suffix = " [log10 pretransform]" if var in pretransform_log10_vars else ""
    ax.set_ylabel(ylab)
    ax.set_xlabel(f"{labels_map.get(var, var)}{prelog_suffix}{xlab_suffix}")
    ax.set_title(f"Distribution of {var} across scenarios — Overlaid (normalized)")

    # NOTE: once we scale, x is on a transformed scale; do not set log xscale here
    # (Log scaling was handled via pretransform_log10_vars)

    legend_elems = [Line2D([0],[0], color=palette.get(s),
                           linestyle=_safe_linestyle(s), linewidth=2.0, label=s)
                    for s in present]
    ax.legend(handles=legend_elems, frameon=False, ncol=2)
    fig.tight_layout()

    base = OUTPUT_DIR / f"dist_{var}_overlaid_norm-{normalize_mode}_scale-{scale_mode}"
    for ext in SAVE_FORMATS:
        fig.savefig(f"{base}.{ext}", dpi=DPI, transparent=TRANSPARENT, bbox_inches="tight")
    plt.close(fig)

# ===================== (Optional) sanity check =====================
# Example: check normalization for the first scenario/variable
# Only runs if there is at least one scenario and variable present.
try:
    scen0 = present[0]
    var0 = vars_to_plot[0]
    v0 = df.loc[df["scenario"] == scen0, var0].to_numpy()
    v0 = v0[np.isfinite(v0)]
    if upper_pct is not None:
        cap0 = np.nanpercentile(df[var0].to_numpy(), upper_pct)
        v0 = v0[v0 <= cap0]
    v0 = pretransform(v0, var0)
    t0, _ = fit_scaler(pretransform(df[var0].to_numpy(), var0), scale_mode)
    v0 = t0(v0)

    b0 = common_bins(v0)
    if normalize_mode == "pdf":
        dens, edges = np.histogram(v0, bins=b0, density=True)
        area = np.sum(dens * np.diff(edges))
        print(f"[Sanity] Area under PDF histogram for {scen0}/{var0}: {area:.3f} (≈1)")
    else:
        dens, edges = np.histogram(v0, bins=b0, weights=np.ones_like(v0)/v0.size)
        height_sum = dens.sum()
        print(f"[Sanity] Sum of fractional heights for {scen0}/{var0}: {height_sum:.3f} (≈1)")
except Exception as e:
    print("[Sanity] Skipped:", e)
```
