# FWI 3-Panel: GFDL-ESM4 vs MRI-ESM2-0 (Avg RH) vs MRI-ESM2-0 (Min RH)

Loads pre-computed annual FWI means from each model's cache (output of FWI_Compute).

```python
from pathlib import Path
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import pandas as pd
```

## 1  Cache paths & scenarios

```python
GFDL_CACHE    = Path("/Users/thunthita/ClimateChangeonWildfire/new/src/GFDL_ESM4/computed_means_GFDL-ESM4")
HURS_CACHE    = Path("/Users/thunthita/ClimateChangeonWildfire/new/src/MRI-ESM2/hurs/computed_means_MRI-ESM2-0")
HURSMIN_CACHE = Path("/Users/thunthita/ClimateChangeonWildfire/new/src/MRI-ESM2/hursmin/computed_means_MRI-ESM2-0_hursmin")

GFDL_SCENARIOS    = ["historical", "ssp119", "ssp126", "ssp245", "ssp370", "ssp585"]
HURS_SCENARIOS    = ["historical", "ssp119", "ssp126", "ssp245", "ssp370", "ssp434", "ssp460", "ssp534", "ssp585"]
HURSMIN_SCENARIOS = ["historical", "ssp119", "ssp126", "ssp245", "ssp370", "ssp460", "ssp534", "ssp585"]
```

## 2  Load from cache

```python
def load_fwi_cache(cache_dir: Path, scenarios: list) -> dict:
    """Load pre-computed FWI annual means. Returns {scenario: DataArray(year)}."""
    out = {}
    for s in scenarios:
        f = cache_dir / f"fwi_{s}.nc"
        if f.exists():
            out[s] = xr.open_dataarray(f).load()
        else:
            print(f"  [warn] missing: {f.name}")
    return out

gfdl_fwi    = load_fwi_cache(GFDL_CACHE,    GFDL_SCENARIOS)
hurs_fwi    = load_fwi_cache(HURS_CACHE,    HURS_SCENARIOS)
hursmin_fwi = load_fwi_cache(HURSMIN_CACHE, HURSMIN_SCENARIOS)

print("GFDL scenarios:    ", list(gfdl_fwi.keys()))
print("hurs scenarios:    ", list(hurs_fwi.keys()))
print("hursmin scenarios: ", list(hursmin_fwi.keys()))
```

## 3  Palette & labels

```python
PALETTE = {
    "historical": "#000000",
    "ssp119":     "#E69F00",
    "ssp126":     "#56B4E9",
    "ssp245":     "#009E73",
    "ssp370":     "#882255",
    "ssp434":     "#E6D600",
    "ssp460":     "#D55E00",
    "ssp534":     "#0072B2",
    "ssp585":     "#CC79A7",
}

SHORT_LABELS = {
    "historical": "Historical",
    "ssp119": "SSP1-1.9", "ssp126": "SSP1-2.6",
    "ssp245": "SSP2-4.5", "ssp370": "SSP3-7.0",
    "ssp434": "SSP4-3.4", "ssp460": "SSP4-6.0",
    "ssp534": "SSP5-3.4OS", "ssp585": "SSP5-8.5",
}

SCENARIO_ORDER = ["historical", "ssp119", "ssp126", "ssp245", "ssp370", "ssp434", "ssp460", "ssp534", "ssp585"]
```

## 4  Plot

```python
def plot_fwi_3panel(
    gfdl_fwi:    dict,
    hurs_fwi:    dict,
    hursmin_fwi: dict,
    smooth_window: int  = 10,
    output_path:   str  = "Pict/FWI_3panel_all_models.png",
) -> None:
    """
    2×2 figure (Elsevier double-column 190 mm):
      [0,0] GFDL-ESM4        [0,1] MRI-ESM2-0 Avg RH (hurs)
      [1,0] MRI-ESM2-0 Min RH (hursmin)   [1,1] Legend
    """
    Path(output_path).parent.mkdir(parents=True, exist_ok=True)

    def _plot_one(ax, fwi_dict, title):
        for scen in SCENARIO_ORDER:
            if scen not in fwi_dict:
                continue
            da = fwi_dict[scen]
            years = da["year"].values
            vals  = da.values
            # rolling smooth
            if smooth_window and smooth_window > 1:
                vals = pd.Series(vals).rolling(smooth_window, center=True, min_periods=1).mean().values
            ax.plot(years, vals, label=SHORT_LABELS.get(scen, scen),
                    color=PALETTE.get(scen, "#333333"), linewidth=1.0, alpha=0.9)
        ax.set_title(title, loc="left", fontsize=9)
        ax.set_xlim(1850, 2100)
        ax.set_ylim(bottom=15)
        ax.set_xticks([1850, 1900, 1950, 2000, 2050, 2100])
        ax.tick_params(labelsize=8)
        ax.grid(True, alpha=0.4)

    fig, axes = plt.subplots(2, 2, figsize=(7.48, 4.0), sharey=True)
    axG, axH, axM, axL = axes[0,0], axes[0,1], axes[1,0], axes[1,1]

    _plot_one(axG, gfdl_fwi,    "a) GFDL-ESM4 Avg RH")
    _plot_one(axH, hurs_fwi,    "b) MRI-ESM2-0 Avg RH")
    _plot_one(axM, hursmin_fwi, "c) MRI-ESM2-0 Min RH")

    axG.set_ylabel("FWI", fontsize=9)
    axM.set_ylabel("FWI", fontsize=9)

    # Legend panel
    axL.axis("off")
    all_scens = list(dict.fromkeys(
        [s for s in SCENARIO_ORDER if s in gfdl_fwi or s in hurs_fwi or s in hursmin_fwi]
    ))
    handles = [plt.Line2D([0],[0], color=PALETTE.get(s,"#333"), linewidth=1.5) for s in all_scens]
    labels  = [SHORT_LABELS.get(s, s) for s in all_scens]
    axL.legend(handles, labels, loc="upper left", frameon=False, fontsize=8, ncol=1)

    plt.tight_layout()
    plt.savefig(output_path, dpi=300, bbox_inches="tight")
    print(f"Saved → {output_path}")
    plt.show()


plot_fwi_3panel(
    gfdl_fwi    = gfdl_fwi,
    hurs_fwi    = hurs_fwi,
    hursmin_fwi = hursmin_fwi,
    smooth_window = 10,
    output_path   = "Pict/FWI_3panel_all_models.png",
)
```
