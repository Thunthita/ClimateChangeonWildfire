```python
import cftime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- Helpers ---
def _scenario_key_from_label(label: str):
    l = label.replace(" ", "")
    if "Historical" in label: return "Historical"
    if "SSP1-1.9" in label or "SSP119" in l: return "SSP119"
    if "SSP1-2.6" in label or "SSP126" in l: return "SSP126"
    if "SSP2-4.5" in label or "SSP245" in l: return "SSP245"
    if "SSP3-7.0" in label or "SSP370" in l: return "SSP370"
    if "SSP5-8.5" in label or "SSP585" in l: return "SSP585"
    return None

def _ten_year_window_days(years=10):
    # Use astronomical year length to keep things consistent
    return int(round(365.2422 * years))

def _tas_to_units(values, in_units: str | None, out_units: str = "C"):
    """Convert tasmax values between Kelvin and Celsius."""
    out_units = out_units.upper()
    if in_units is None:
        in_units_norm = "K"  # CMIP default
    else:
        u = in_units.strip().lower().replace("°", "").replace("deg", "")
        if u in {"k", "kelvin"}: in_units_norm = "K"
        elif u in {"c", "celsius"}: in_units_norm = "C"
        else: in_units_norm = "K"  # safe default

    if in_units_norm == "K" and out_units == "C":
        return values - 273.15
    if in_units_norm == "C" and out_units == "K":
        return values + 273.15
    return values  # same unit
```

```python
def _to_mm_per_year(pr_values, pr_units: str | None):
    """
    Convert precipitation to mm/yr.
    Common cases:
      - 'kg m-2 s-1' (a.k.a. mm/s)  -> * 86400 * 365.2422
      - 'mm/day'                    -> * 365.2422
      - 'mm/yr'                     -> * 1
    If units are unknown, we assume 'kg m-2 s-1' (CMIP convention).
    """
    DAYS_PER_YEAR = 365.2422
    SECONDS_PER_DAY = 86400.0

    if pr_units is None:
        # Fallback assumption (CMIP-style flux): kg m-2 s-1
        return pr_values * SECONDS_PER_DAY * DAYS_PER_YEAR

    u = pr_units.strip().lower()
    if "kg" in u and "s-1" in u:
        return pr_values * SECONDS_PER_DAY * DAYS_PER_YEAR
    if "mm/day" in u or "mm d-1" in u:
        return pr_values * DAYS_PER_YEAR
    if "mm/yr" in u or "mm year-1" in u:
        return pr_values
    # Unknown unit → safest common case: kg m-2 s-1
    return pr_values * SECONDS_PER_DAY * DAYS_PER_YEA
```

```python
# Define Earth's radius in meters
R = 6378137

# Latitude and longitude arrays (in degrees)
lon_tas = datasets_tas_historical.coords['lon']
lat_tas = datasets_tas_historical.coords['lat']

# Convert to radians
lon_radians_tas = np.radians(lon_tas)
lat_radians_tas = np.radians(lat_tas)

# Calculate the differences in longitude and latitude
delta_lon_tas = np.diff(lon_radians_tas)[0]  # Assuming regular grid
delta_lat_tas = np.diff(lat_radians_tas).mean()  # Take the mean for approximate spacing

# Calculate area for each grid cell
area_tas = np.zeros((len(lat_tas), len(lon_tas)))  # Initialize area array to match full grid

for i in range(len(lat_tas)):
    # Latitude bounds (extend to midpoints for full grid)
    if i == 0:  # Southernmost edge
        phi1_tas = lat_radians_tas[i] - delta_lat_tas / 2
        phi2_tas = lat_radians_tas[i] + delta_lat_tas / 2
    elif i == len(lat_tas) - 1:  # Northernmost edge
        phi1_tas = lat_radians_tas[i] - delta_lat_tas / 2
        phi2_tas = lat_radians_tas[i] + delta_lat_tas / 2
    else:  # Interior latitudes
        phi1_tas = lat_radians_tas[i] - delta_lat_tas / 2
        phi2_tas = lat_radians_tas[i] + delta_lat_tas / 2

    # Calculate area for this latitude band
    area_tas[i, :] = (R**2 * delta_lon_tas * (np.sin(phi2_tas) - np.sin(phi1_tas)))

# Convert to square kilometers
area_km2_tas = area_tas / 1e6

# Print or save the area array
print("Area array (in km^2):", area_km2_tas)
print("Shape of area array:", area_km2_tas.shape)
```
