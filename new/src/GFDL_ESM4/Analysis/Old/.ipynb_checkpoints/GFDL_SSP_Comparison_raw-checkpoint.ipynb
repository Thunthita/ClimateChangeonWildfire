# SSP Scenario FWI Comparison — GFDL-ESM4 (Raw)
Compares Fire Weather Index (FWI) between SSP scenarios using raw NetCDF files.
Time slices: **2040–2050** and **2090–2100**.

## 1. Imports

```python
import numpy as np
import xarray as xr
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

## 2. Load Datasets

```python
# File paths
BASE = '/Volumes/jubjang/Jubjang/FWICMIP6'
file_SSP119 = f'{BASE}/fwisa_ann_GFDL-ESM4_ssp119_r1i1p1f1_g025.nc'
file_SSP370 = f'{BASE}/fwisa_ann_GFDL-ESM4_ssp370_r1i1p1f1_g025.nc'
file_SSP585 = f'{BASE}/fwisa_ann_GFDL-ESM4_ssp585_r1i1p1f1_g025.nc'

dataset_SSP119 = xr.open_mfdataset(file_SSP119, combine='by_coords', engine='netcdf4')
dataset_SSP370 = xr.open_mfdataset(file_SSP370, combine='by_coords', engine='netcdf4')
dataset_SSP585 = xr.open_mfdataset(file_SSP585, combine='by_coords', engine='netcdf4')

print('Datasets loaded.')
```

## 3. Define Area-Weighted Grid for Thailand

```python
# Compute grid-cell areas (km²) using spherical geometry
R = 6_378_137  # Earth radius (m)
num_lat, num_lon = 72, 144
lat = np.linspace(-88.75, 88.75, num_lat)
lon = np.linspace(1.25, 358.75, num_lon)

lat_rad   = np.radians(lat)
delta_lon = np.radians(np.diff(lon)[0])
delta_lat = np.radians(np.diff(lat).mean())

area = np.zeros((num_lat, num_lon))
for i in range(num_lat):
    phi1 = lat_rad[i] - delta_lat / 2
    phi2 = lat_rad[i] + delta_lat / 2
    area[i, :] = R**2 * delta_lon * (np.sin(phi2) - np.sin(phi1))

area_km2 = area / 1e6

areacella = xr.Dataset({
    'area': xr.DataArray(
        area_km2,
        dims=['lat', 'lon'],
        coords={'lat': lat, 'lon': lon},
    )
})

thailand_lat_range = slice(5.5, 20.5)
thailand_lon_range = slice(97.5, 105.5)
areacella_th = areacella.sel(lat=thailand_lat_range, lon=thailand_lon_range)
print('areacella_th shape:', areacella_th['area'].shape)
```

## 4. Extract FWI Time Slices

```python
def calculate_fwi_statistics(dataset, areacella_th):
    """
    Extract area-weighted annual mean FWI over Thailand for two decades:
    2040-2050 and 2090-2100.
    Returns two np.ndarrays.
    """
    fwi_th = dataset['fwisa'].sel(
        lat=thailand_lat_range, lon=thailand_lon_range
    )
    weights = areacella_th['area']
    fwi_weighted = fwi_th.weighted(weights).mean(dim=['lat', 'lon'])

    fwi_2040_2050 = fwi_weighted.sel(time=slice('2040', '2050')).values
    fwi_2090_2100 = fwi_weighted.sel(time=slice('2090', '2100')).values
    return fwi_2040_2050, fwi_2090_2100


fwi_2040_2050_SSP119, fwi_2090_2100_SSP119 = calculate_fwi_statistics(dataset_SSP119, areacella_th)
fwi_2040_2050_SSP370, fwi_2090_2100_SSP370 = calculate_fwi_statistics(dataset_SSP370, areacella_th)
fwi_2040_2050_SSP585, fwi_2090_2100_SSP585 = calculate_fwi_statistics(dataset_SSP585, areacella_th)

print('SSP119 2040-2050 n =', len(fwi_2040_2050_SSP119))
print('SSP119 2090-2100 n =', len(fwi_2090_2100_SSP119))
```

## 5. Helper Functions

```python
def cohens_d(x1, x2):
    """Pooled Cohen's d effect size (positive = x2 > x1)."""
    n1, n2 = len(x1), len(x2)
    s_pooled = np.sqrt(
        ((n1 - 1) * np.var(x1, ddof=1) + (n2 - 1) * np.var(x2, ddof=1))
        / (n1 + n2 - 2)
    )
    return (np.mean(x2) - np.mean(x1)) / s_pooled


def mean_diff_confidence_interval(a, b, confidence=0.95):
    """95% CI for the difference in means (b - a)."""
    diff = np.mean(b) - np.mean(a)
    se = np.sqrt(np.var(a, ddof=1) / len(a) + np.var(b, ddof=1) / len(b))
    t_crit = stats.t.ppf((1 + confidence) / 2, df=min(len(a), len(b)) - 1)
    margin = t_crit * se
    return diff, diff - margin, diff + margin
```

## 6. SSP Comparison Function

```python
def compare_ssp_same_decade(fwi_ssp1, fwi_ssp2, ssp1_label, ssp2_label,
                            decade, filename, alpha=0.05):
    _, p1 = stats.shapiro(fwi_ssp1)
    _, p2 = stats.shapiro(fwi_ssp2)
    print(f'Shapiro-Wilk p ({ssp1_label}): {p1:.4f}')
    print(f'Shapiro-Wilk p ({ssp2_label}): {p2:.4f}')

    if p1 < alpha or p2 < alpha:
        diff = np.array(fwi_ssp2) - np.array(fwi_ssp1)
        if np.all(diff == 0):
            p_value, test_name = 1.0, 'Wilcoxon (no variation)'
        else:
            _, p_value = stats.wilcoxon(fwi_ssp2, fwi_ssp1, alternative='greater')
            test_name = 'Wilcoxon Signed-Rank Test'
    else:
        _, p_value = stats.ttest_rel(fwi_ssp2, fwi_ssp1, alternative='greater')
        test_name = 'One-tailed Paired t-Test'

    print(f'\nTest: {test_name}')
    print(f'p-value: {p_value:.4f}')
    if p_value < alpha:
        print(f'Reject H0 — FWI of {ssp2_label} > {ssp1_label}.')
    else:
        print(f'Fail to reject H0.')

    diff_val, ci_lo, ci_hi = mean_diff_confidence_interval(fwi_ssp1, fwi_ssp2)
    d_val = cohens_d(fwi_ssp1, fwi_ssp2)
    effect = 'Very small' if abs(d_val)<0.2 else 'Small' if abs(d_val)<0.5 else 'Medium' if abs(d_val)<0.8 else 'Large'
    print(f'Mean {ssp1_label}: {np.mean(fwi_ssp1):.2f}')
    print(f'Mean {ssp2_label}: {np.mean(fwi_ssp2):.2f}')
    print(f'Diff: {diff_val:.2f}  95% CI: ({ci_lo:.2f}, {ci_hi:.2f})')
    print(f"Cohen's d: {d_val:.2f} ({effect})")

    fig, ax = plt.subplots(figsize=(5, 4))
    ax.boxplot([fwi_ssp1, fwi_ssp2], labels=[ssp1_label, ssp2_label])
    ax.set_ylabel('FWI')
    ax.set_title(f'FWI Comparison ({decade})\n{test_name}')
    ax.grid(True, linestyle='--', alpha=0.5)
    fig.tight_layout()
    fig.savefig(filename + f'_SSP_Comparison_{decade}.png', dpi=300)
    plt.show()
    print('-' * 60)
```

## 7. Run Comparisons

```python
compare_ssp_same_decade(
    fwi_2040_2050_SSP119, fwi_2040_2050_SSP585,
    ssp1_label='SSP1-1.9', ssp2_label='SSP5-8.5',
    decade='2040–2050', filename='fwi_comparison_2040s_119_585'
)
```

```python
compare_ssp_same_decade(
    fwi_2040_2050_SSP119, fwi_2040_2050_SSP370,
    ssp1_label='SSP1-1.9', ssp2_label='SSP3-7.0',
    decade='2040–2050', filename='fwi_comparison_2040s_119_370'
)
```

```python
compare_ssp_same_decade(
    fwi_2090_2100_SSP119, fwi_2090_2100_SSP585,
    ssp1_label='SSP1-1.9', ssp2_label='SSP5-8.5',
    decade='2090–2100', filename='fwi_comparison_2090s_119_585'
)
```

```python
compare_ssp_same_decade(
    fwi_2090_2100_SSP119, fwi_2090_2100_SSP370,
    ssp1_label='SSP1-1.9', ssp2_label='SSP3-7.0',
    decade='2090–2100', filename='fwi_comparison_2090s_119_370'
)
```
